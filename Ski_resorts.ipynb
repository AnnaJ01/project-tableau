```python
def multiply(row):
 
    if row['currency'] == 'Czech koruna':
        return row['adult'] * 0.041
    elif row['currency'] == 'Norwegian krone':
        return row['adult'] * 0.088
    elif row['currency'] == 'Polish zloty':
        return row['adult'] * 0.24
    elif (row['currency'] == 'Swedish krona') & (row['adult'] > 80):
        return row['adult'] * 0.087
    elif row['currency'] == 'UK Pound':
        return row['adult'] * 1.18
    elif row['currency'] == 'Bulgarian Lev':
        return row['adult'] * 0.51
    elif row['currency'] == 'Romanian leu':
        return row['adult'] * 0.20
    elif row['currency'] == 'Bosnia convertible mark':  
        return row['adult'] * 0.51
    elif row['currency'] == 'Icelandic krona':
        return row['adult'] * 0.0067
    elif row['currency'] == 'Hungarian forint':
        return row['adult'] * 0.0026
    elif row['currency'] == 'Georgian Lari':
        return row['adult'] * 0.34
    elif row['currency'] == 'Turkish lira':
        return row['adult'] * 0.029
    elif row['currency'] == 'Swiss Franc':
        return row['adult'] * 1.01
    elif row['currency'] == 'Azerbaijani manat':
        return row['adult'] * 0.54
    elif row['currency'] == 'Serbian dinar':   
        return row['adult'] * 0.0085
    elif row['currency'] == 'Ukrainian hryvnia':
        return row['adult'] * 0.023
    elif row['currency'] == 'Macedonian denar':
        return row['adult'] * 0.016
    elif row['currency'] == 'Danish Krone':
        return row['adult'] * 0.13
#    elif row['Currency'] == 'Croatian Kuna':
#       return row['Adult'] * 0.1327
    else:
        return row['adult']

        
df_eu['adult'] = df_eu.apply(multiply, axis=1)
```

```python

#show the slopes difficulty

grouped = df_eu.groupby('country')[['easy', 'intermediate_', 'difficult']].sum()

# Plot the grouped bar chart
plt.figure(figsize=(12, 8))
barWidth = 0.25

# Set the position of the bars on the x-axis
r1 = np.arange(len(grouped.index))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# Create the bars
plt.bar(r1, grouped['easy'], color='b', width=barWidth, edgecolor='grey', label='Beginner')
plt.bar(r2, grouped['intermediate_'], color='g', width=barWidth, edgecolor='grey', label='Intermediate')
plt.bar(r3, grouped['difficult'], color='r', width=barWidth, edgecolor='grey', label='Difficult')

# Add labels
plt.xlabel('Country', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(grouped.index))], grouped.index, rotation=90)
plt.ylabel('Count')
plt.title('Total Number of Slopes by Country and Level')
plt.legend()
plt.show()
```

```python
# Sort the DataFrame by altitude in descending order
df_eu_sorted = df_eu.sort_values(by='altitude', ascending=False).head(20)

# Create a bar plot to show the altitude of the top ten highest resorts
plt.figure(figsize=(12, 8))
plt.bar(df_eu_sorted['nameresort'], df_eu_sorted['altitude'], color='skyblue')
plt.xlabel('Ski Resort')
plt.ylabel('Altitude (m)')
plt.title('Top Ten Highest Altitude Ski Resorts in Europe')
plt.xticks(rotation=45, ha='right')
plt.show()
```
